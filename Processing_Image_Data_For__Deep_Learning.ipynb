# **Resizing the image using 'Pillow library'**

```python
from PIL import Image  # Image is the module
```

```python
img = Image.open('/content/cat.jpg')
img_resized = img.resize((200, 200))  # Mentioning the dimension , Basically reducing the dimension of the image.

# Saving this 'Resize image'
```

```python
img_resized.save('cat_resized.jpg') # Have to mention the file name that want to save.
```

```python
# Displaying the image from numpy array

img_res = mpimg.imread('/content/cat_resized.jpg')  # This image will be read and store in this variable called 'img'.
img_res_plot = plt.imshow(img_res)
plt.show()
```

# **Convert RGB images to Grayscale image using OpenCV**

```python
# importing OpenCV Library

import cv2
```

```python
img = cv2.imread('/content/cat.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```

***The Grayscale image have only 1 channel..that's why we don't have the value of 3***

```python
from google.colab.patches import cv2_imshow
```

```python
# Display the image (grayscale image)
cv2_imshow(grayscale_image)
```

```python
# Saving the grayscale image
cv2.imwrite('cat_grayscale.jpg', grayscale_image)
```

# Properties of Grayscale image:
*   All pixel value lies between 0 and 255
*   0 means black & 255 means white
*   1 Channel

 #### Converting an ***RGB image to a grayscale image*** significantly reduces the amount of memory required because, instead of ***three matrices representing the Red, Green, and Blue color channels***, we only need ***one matrix***. ***This results in a substantial space-saving and can improve processing speed, especially beneficial for applications that don't need color information, like certain machine learning models or image analysis tasks.***
